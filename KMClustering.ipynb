```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dts= pd.read_csv('C:/Users/saswa/saswata/Machine Learning A-Z (Codes and Datasets)/Part 4 - Clustering/Section 24 - K-Means Clustering/Python/Mall_Customers.csv')
```

```python
x= dts.iloc[:, :].values                  #.values convert the pd frame into nd array. Here for the clusterin, we find pattern in all the features
x_2d= dts.iloc[:, 3:].values               #here i chose 2 dimentional input for easy clustering visualization
```

```python
from sklearn.cluster import KMeans
```

```python
wcss= list()
for i in range(1,11):
    kmeans= KMeans(n_clusters=i, init= 'k-means++', random_state= 42)
    kmeans.fit(x_2d)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)    
```

```python
kmeans= KMeans(n_clusters=5, init= 'k-means++', random_state= 42)
y_kmeans= kmeans.fit_predict(x_2d)
```

```python
print(y_kmeans)             #shows each respective customers in the cluster number 0-4
```

```python
plt.scatter(x_2d[y_kmeans==0, 0], x_2d[y_kmeans==0, 1], s=80, c= 'red', label= 'Cluster 1')
plt.scatter(x_2d[y_kmeans==1, 0], x_2d[y_kmeans==1, 1], s=80, c= 'blue', label= 'Cluster 2')
plt.scatter(x_2d[y_kmeans==2, 0], x_2d[y_kmeans==2, 1], s=80, c= 'black', label= 'Cluster 3')
plt.scatter(x_2d[y_kmeans==3, 0], x_2d[y_kmeans==3, 1], s=80, c= 'green', label= 'Cluster 4')
plt.scatter(x_2d[y_kmeans==4, 0], x_2d[y_kmeans==4, 1], s=80, c= 'violet', label= 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=80, c= 'yellow', label= 'Centroid')         #s signifies the size of that scatter dot
plt.legend()
plt.show()
```
